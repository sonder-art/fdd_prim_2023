## Tipos de Datos

Realiza las transformaciones o casteos (casting) que creas necesarios a tus datos de tal manera que el typo de dato sea adecuado. Al terminar recrea la tabla `column_types` con los nuevos tipos.

```python
# Revisamos la cnatidad de NaN's en la tabla
loans.isna().sum()
```

```python
# Establecemos la columna id como el √≠ndice de la tabla
loans.set_index("id", inplace=True)
```

```python
# Hacemos una partici√≥n del dataset original solo con las columnas que pueden ser de interes para el an√°lisis que se va a realizar. Decid√≠ quedarme con las variables que ten√≠an m√°s datos y que, basado en su descripci√≥n pod√≠an ser √∫tiles.
loans_subset = loans[['loan_amnt' , 'funded_amnt', 'funded_amnt_inv', 'term', 'int_rate', 'installment', 'grade', 'sub_grade', 'emp_length', 'home_ownership', 'annual_inc', 'verification_status', 'loan_status', 'pymnt_plan', 'purpose', 'dti', 'delinq_2yrs', 'earliest_cr_line', 'open_acc', 'revol_bal', 'total_acc', 'total_pymnt', 'total_pymnt_inv', 'last_pymnt_d', 'last_pymnt_amnt', 'tot_cur_bal', 'acc_open_past_24mths', 'avg_cur_bal', 'pct_tl_nvr_dlq', 'tot_hi_cred_lim']]
loans_subset
```

```python
# Manejos de tipos 1
# Tu codigo aqui
# De la columna term reemplazamos los valores por enteros y cambiamos el tipo de dato de la columna a int para facilitar su an√°lisis
loans_subset["term"].replace({" 36 months":"36", " 60 months":"60"}, inplace=True)
loans_subset["term"] = loans_subset["term"].astype("int")
loans_subset
```

```python
# Se pasa todo lo que sea string a min√∫sculas
loans_subset = loans_subset.applymap(lambda s: s.lower() if type(s) == str else s)
loans_subset.head()
```

```python
# Manejos de tipos 2
# Tu codigo aqui
# Pasamos a Datetime las variables que son fehcas
loans_subset['last_pymnt_d'] = pd.to_datetime(loans_subset['last_pymnt_d'])
loans_subset['last_pymnt_d'] = loans_subset['last_pymnt_d'].dt.to_period('M')
loans_subset['last_pymnt_d']
```

```python
loans_subset['earliest_cr_line'] = pd.to_datetime(loans_subset['earliest_cr_line'])
loans_subset['earliest_cr_line'] = loans_subset['earliest_cr_line'].dt.to_period('M')
loans_subset['earliest_cr_line']
```

```python
# Reemplazamos los datos de la variable "emp_length" para poder pasar el tipo de dato de la columna a float
loans_subset["emp_length"].replace({"10+ years":"10", "8 years":"8", "6 years":"6", "< 1 year":"0", "2 years":"2"
, "9 years":"9", "7 years":"7", "5 years":"5", "3 years":"3", "1 year":"1", "4 years":"4"}, inplace=True)
loans_subset["emp_length"] = loans_subset["emp_length"].astype("float")
loans_subset.head()
```

```python
# Establecemos la variable "pymnt_plan" como boolean reemplazando sus valores para facilitar el an√°lisis
loans_subset["pymnt_plan"].replace({"n":False, "y":True}, inplace=True)
loans_subset.head()
```

```python
# Revisamos los nuevos tipos de datos de las columnas 
column_types = loans_subset.dtypes
column_types
```

## Manejo de NaNs o missings

Maneja los datos de tipos missing. Elije una estrategia adecuada dependiendo del tipo de dato que le asignaste a la columna.


Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. Por ejemplo: Si hay una columna que se llama `columna 3` y utilizaste la estrategia de imputacion de media, y existe otra llamada `columna 4` y  elegiste la palabra 'missing' el JSON debera contener:  
  
 `{'columna 3':{'estrategia':'mean', 'valor':3.4}, 'columna 4':{'estrategia':'identificador', 'valor':'missing'}}`  

 De tal manera que para cada columna que tenga un metodo de imputacion apunte a otro diccionario donde el **key** `estrategia` describa de manera sencilla el metodo, y el **key** `valor` el valor usado. En general:   
 `{'nombre de la columna':{'estrategia':'descripcion de estrategia', 'valor':'valor utilizado'}}`. 
 

De utilizar mas de un metodo puedes anidarlos en una lista  
  `[{...},{...}]`.  

Incluso si la columna utilizada no sufrio imputacion, es necesario que la agregues al JSON.

La idea es que cualquier otra persona pueda cargar el el archivo JSON con tu funcion, entender que hiciste y replicarlo facilmente. No existe solo una respuesta correcta, pero tendras que justificar y explicar tus deciciones.

### Imputacion

```python
# Volvemos a revisar cu√°ntos NaN's tenemos
loans_subset.isna().sum()
```

```python
# Manejamos los NaN's de la columna "last_pymnt_d" poniendo un "missing" ya que como se trata de un fecha es m√°s dif√≠cil poner alg√∫n valor
loans_subset["last_pymnt_d"].fillna("missing", inplace=True)
```

```python
# Como la columna "emp_length" guarda los a√±os de trabajo de la persona, podemos aplicar una imputaci√≥n anal√≠tica
# En este caso opt√© por el promedio para no perder los dem√°s datos y poder realizar el an√°lisis de la variable 
avg_emp_length = np.mean(loans_subset["emp_length"])  
loans_subset["emp_length"].fillna(avg_emp_length, inplace=True)
loans_subset.head()
```

```python
# Revisamos que ya no queden NaN's
loans_subset.isna().sum()
```
