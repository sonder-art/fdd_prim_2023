## Manejo de NaNs o missings

Maneja los datos de tipos missing. Elije una estrategia adecuada dependiendo del tipo de dato que le asignaste a la columna.


Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. Por ejemplo: Si hay una columna que se llama `columna 3` y utilizaste la estrategia de imputacion de media, y existe otra llamada `columna 4` y  elegiste la palabra 'missing' el JSON debera contener:  
  
 `{'columna 3':{'estrategia':'mean', 'valor':3.4}, 'columna 4':{'estrategia':'identificador', 'valor':'missing'}}`  

 De tal manera que para cada columna que tenga un metodo de imputacion apunte a otro diccionario donde el **key** `estrategia` describa de manera sencilla el metodo, y el **key** `valor` el valor usado. En general:   
 `{'nombre de la columna':{'estrategia':'descripcion de estrategia', 'valor':'valor utilizado'}}`. 
 

De utilizar mas de un metodo puedes anidarlos en una lista  
  `[{...},{...}]`.  

Incluso si la columna utilizada no sufrio imputacion, es necesario que la agregues al JSON.

La idea es que cualquier otra persona pueda cargar el el archivo JSON con tu funcion, entender que hiciste y replicarlo facilmente. No existe solo una respuesta correcta, pero tendras que justificar y explicar tus deciciones.

### Imputacion

```python
columnas_With_Nans = loans_subset.isna().sum()
columnas_With_Nans
```

```python
columnas_With_Nans[columnas_With_Nans >0]
```

```python
loans_subset = loans_subset.drop(['hardship_amount','hardship_length','hardship_dpd', 'deferral_term','orig_projected_additional_accrued_interest','hardship_payoff_balance_amount', 'hardship_last_payment_amount','settlement_amount', 'settlement_percentage','settlement_term'], axis = 1)
```

```python
loans_subset['emp_title'] = loans_subset['emp_title'].fillna("Unknown")
```

```python
loans_subset['emp_length'].fillna("Unknown years",inplace=True)
```

```python
loans_subset['mths_since_last_delinq'] = loans_subset['mths_since_last_delinq'].fillna(" Not applicable")
loans_subset['mths_since_last_record'] = loans_subset['mths_since_last_record'].fillna(" Not applicable")
```

```python
loans_subset['revol_util'].fillna(loans_subset['revol_util'].mean(), inplace=True)
loans_subset['mo_sin_old_il_acct'].fillna(loans_subset['mo_sin_old_il_acct'].mean(), inplace=True)
loans_subset['mths_since_recent_bc'].fillna(loans_subset['mths_since_recent_bc'].mean(), inplace=True)
loans_subset['mths_since_recent_bc_dlq'].fillna(loans_subset['mths_since_recent_bc_dlq'].mean(), inplace=True)
loans_subset['mths_since_recent_inq'].fillna(loans_subset['mths_since_recent_inq'].mean(), inplace=True)
loans_subset['mths_since_recent_revol_delinq'].fillna(loans_subset['mths_since_recent_revol_delinq'].mean(), inplace=True)
loans_subset['mths_since_last_major_derog'].fillna(loans_subset['mths_since_last_major_derog'].mean(), inplace=True)
loans_subset['bc_util'].fillna(loans_subset['bc_util'].mean(), inplace=True)
loans_subset['percent_bc_gt_75'].fillna(loans_subset['percent_bc_gt_75'].mean(), inplace=True)
```

```python
loans_subset['bc_open_to_buy'].fillna(loans_subset['bc_open_to_buy'].mode()[0], inplace=True)
loans_subset['num_tl_120dpd_2m'].fillna(loans_subset['num_tl_120dpd_2m'].mode()[0], inplace=True)
loans_subset['last_pymnt_d'].fillna(loans_subset['last_pymnt_d'].mode()[0], inplace=True)
loans_subset['next_pymnt_d'].fillna(loans_subset['next_pymnt_d'].mode()[0], inplace=True)
loans_subset['last_credit_pull_d'].fillna(loans_subset['last_credit_pull_d'].mode()[0], inplace=True)
```

```python
columnas_With_Nans = loans_subset.isna().sum()
columnas_With_Nans[columnas_With_Nans >0]
```
