### Imputacion

```python
#Primero verificamos que columnas tienen gran nmero de Nans
#en este caso los que sean ms de 10000
suma_nans = loans.isna().sum()
columnas_selecc = suma_nans.between(10000, 90000)
columnas_con_nans = loans.loc[:, columnas_selecc]
columnas_sin_datos = loans.loc[:, suma_nans > 90000]
print(columnas_con_nans.columns)
print(columnas_sin_datos.columns)
```

```python
#Luego verificamos si la distribucin es normal o sesgada
columnas_con_nans.hist(figsize=(100, 100))
plt.show()
```

Podemos ver que la variable columnas_sin_datos contiene ms de 90 mil Nans, por lo que se eliminarn 

Adicionalmente la variable columnas_con_nans almacena las columnas a las que se les aplicar una estrategia de imputacin de mediana puesto que la distribucin es sesgada
        

```python
# Tu codigo aqui
imputacion = loans.drop(columns=columnas_sin_datos.columns)
imputacion = imputacion.fillna(imputacion.median())
```
